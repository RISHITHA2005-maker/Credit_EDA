```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
app_df = pd.read_csv("application_data.csv") 
```

```python
print("Application data shape:", app_df.shape)
```

```python
display(app_df.head())
display(app_df.info())
display(app_df.describe(include='all'))
```

```python
missing_vals = app_df.isnull().sum()
missing_percent = (missing_vals / len(app_df)) * 100
missing_df = pd.DataFrame({'Missing Values': missing_vals, 'Percentage': missing_percent})
display(missing_df[missing_df['Missing Values'] > 0].sort_values(by='Percentage', ascending=False))
```

```python
for col in app_df.select_dtypes(include='number'):
    if app_df[col].isnull().any():
        median_val = app_df[col].median()
        app_df[col].fillna(median_val, inplace=True)
```

```python
for col in app_df.select_dtypes(include='object'):
    if app_df[col].isnull().any():
        mode_val = app_df[col].mode()
        if not mode_val.empty:
            app_df[col].fillna(mode_val[0], inplace=True)
        else:
            app_df[col].fillna("Unknown", inplace=True) 
```

```python
print(app_df.dtypes)
```

```python
app_df['AGE_BIN'] = pd.cut(app_df['DAYS_BIRTH'] / -365, bins=[20, 30, 40, 50, 60, 70], 
                           labels=['20s', '30s', '40s', '50s', '60s'])
```

```python
sns.countplot(x='TARGET', data=app_df)
plt.title("Target Variable Distribution")
plt.show()
print(app_df['TARGET'].value_counts(normalize=True))
```

```python
categorical_cols = app_df.select_dtypes(include='object').columns
for col in categorical_cols[:5]:  # Plot first 5 for brevity
    plt.figure(figsize=(8, 4))
    sns.countplot(data=app_df, x=col, order=app_df[col].value_counts().index)
    plt.xticks(rotation=45)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
numerical_cols = app_df.select_dtypes(include='number').columns
app_df[numerical_cols].hist(figsize=(16, 12), bins=30)
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(x='TARGET', y='AMT_INCOME_TOTAL', data=app_df)
plt.title("Income vs Target")
plt.show()
```

```python
sns.pairplot(app_df[['AMT_INCOME_TOTAL', 'AMT_CREDIT', 'AMT_ANNUITY', 'TARGET']].sample(1000), hue='TARGET')
plt.show()
```

```python
corr = app_df[numerical_cols].corr()
plt.figure(figsize=(14, 10))
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap")
plt.show()
```
